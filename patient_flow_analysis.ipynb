```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load hospital patient flow data from CSV file into a pandas DataFrame
df = pd.read_csv("data/sample_hospital_patient_flow_data.csv")
```

```python

# Display the first 100 rows of the dataset to preview patient records
df.head(100)

# Show the shape of the dataset (number of rows, number of columns)
df.shape

# Display the column names in the dataset to understand available data fields
df.columns

# Display the last 5 rows of the dataset to check the end of the data
df.tail()

# Display the column names in the dataset to understand available data fields
df.columns
```

```python
# Count and display the total number of patients (rows) in the dataset
print(f"Total patients in dataset: {len(df)}")
```

```python
# Compute the average waiting time for patients in the Emergency Department
# The .mean() function calculates the average of the 'wait_time_minutes' column
average_waiting_time = df["wait_time_minutes"].mean()

# Print the average waiting time result for analysis and reporting
print("Average Waiting Time:", average_waiting_time)
```

```python
# Calculate the maximum waiting time from the 'wait_time_minutes' column
# This identifies the longest time any patient waited in the Emergency Department
max_waiting_time = df["wait_time_minutes"].max()

# Display the maximum waiting time
print("Maximum Waiting Time:", max_waiting_time)
```

```python
# Filter patients who waited 180 minutes (3 hours) or more
# df["wait_time_minutes"] >= 180 creates a condition to find long waiting cases
# .shape[0] counts the number of patients matching this condition
patients_over_3hrs = df[df["wait_time_minutes"] >= 180].shape[0]

# Display the total number of patients who waited more than 3 hours
print("Patients waiting more than 3 hours:", patients_over_3hrs)
```
